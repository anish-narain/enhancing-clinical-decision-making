```python
import matplotlib.pyplot as plt
import numpy as np

def plot_comparison_bar_charts(original_cp, enhanced_cp, original_lp, enhanced_lp, original_title, enhanced_title, full_title, img_name):
    # Evaluation metrics for the x-axis labels
    metrics = ['Accuracy', 'AUC', 'Precision', 'Recall', 'F1 Score']

    # Number of bars (assuming all arrays have the same length)
    n = len(metrics)

    # Bar width
    bar_width = 0.35

    # Index for the groups
    index = np.arange(n)

    # Create the subplot
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(18, 8))

    # Plotting for concept predictor
    rects1_cp = ax1.bar(index, original_cp, bar_width, label=original_title, color='b')
    rects2_cp = ax1.bar(index + bar_width, enhanced_cp, bar_width, label=enhanced_title, color='g')

    ax1.set_xlabel('Evaluation Metrics', fontsize=14)
    ax1.set_ylabel('Scores', fontsize=14)
    ax1.set_title('Concept Predictor', fontsize=16)
    ax1.set_xticks(index + bar_width / 2)
    ax1.set_xticklabels(metrics, fontsize=12)
    ax1.legend(loc='lower right', fontsize=12)

    # Plotting for label predictor
    rects1_lp = ax2.bar(index, original_lp, bar_width, label=original_title, color='b')
    rects2_lp = ax2.bar(index + bar_width, enhanced_lp, bar_width, label=enhanced_title, color='g')

    ax2.set_xlabel('Evaluation Metrics', fontsize=14)
    ax2.set_ylabel('Scores', fontsize=14)
    ax2.set_title('Label Predictor', fontsize=16)
    ax2.set_xticks(index + bar_width / 2)
    ax2.set_xticklabels(metrics, fontsize=12)
    ax2.legend(loc='lower right', fontsize=12)

    # Set the overall title
    fig.suptitle(full_title, fontsize=18)

    # Function to add labels to the bars
    def autolabel(ax, rects):
        for rect in rects:
            height = rect.get_height()
            ax.annotate(f'{height:.2f}',
                        xy=(rect.get_x() + rect.get_width() / 2, height),
                        xytext=(0, 3),  # 3 points vertical offset
                        textcoords="offset points",
                        ha='center', va='bottom', fontsize=12)

    autolabel(ax1, rects1_cp)
    autolabel(ax1, rects2_cp)
    autolabel(ax2, rects1_lp)
    autolabel(ax2, rects2_lp)

    # Adjust layout
    fig.tight_layout(rect=[0, 0, 1, 0.95])

    # Save the plot
    plt.savefig(img_name)

    # Show the plot
    plt.show()
```

```python
# Example usage
original_cp = [0.85, 0.88, 0.87, 0.86, 0.89]
enhanced_cp = [0.90, 0.91, 0.92, 0.93, 0.94]
original_lp = [0.80, 0.82, 0.81, 0.83, 0.84]
enhanced_lp = [0.85, 0.87, 0.88, 0.89, 0.90]
original_title = 'Original'
enhanced_title = 'Enhanced'
full_title = 'Comparison of Predictors'
img_name = 'comparison_plot.png'

plot_comparison_bar_charts(original_cp, enhanced_cp, original_lp, enhanced_lp, original_title, enhanced_title, full_title, img_name)
```
